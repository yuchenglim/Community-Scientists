```python
import numpy as np
import matplotlib.pyplot as plt
```

RC differentiator of square wave:  
Capacitor is high pass filter  
High frequency components give the edges (parts that change)  
Low frequency components everything else

So we try to plot the sum of the high and low frequency components separately!

```python
low_f = np.linspace(1, 50, num = 50, endpoint = True)
high_f = np.linspace(51, 100, num = 50, endpoint = True)

t = np.linspace(-10, 10, num = 200, endpoint = True)

square_wave = (np.array([numpy.sum(np.sin(low_f*np.pi/2)/(low_f*np.pi/2)*2*np.cos(low_f*np.pi/2*x)) for x in t])+
np.array([numpy.sum(np.sin(high_f*np.pi/2)/(high_f*np.pi/2)*2*np.cos(high_f*np.pi/2*x)) for x in t]) )

plt.plot(t, square_wave)
###################################

low_f = np.linspace(1, 5, num = 5, endpoint = True)

t = np.linspace(-10, 10, num = 200, endpoint = True)

square_wave = np.array([numpy.sum(np.sin(low_f*np.pi/2)/(low_f*np.pi/2)*2*np.cos(low_f*np.pi/2*x)) for x in t])

plt.plot(t, square_wave)


###################################


high_f = np.linspace(51, 100, num = 50, endpoint = True)

t = np.linspace(-10, 10, num = 200, endpoint = True)

square_wave = np.array([numpy.sum(np.sin(high_f*np.pi/2)/(high_f*np.pi/2)*2*np.cos(high_f*np.pi/2*x)) for x in t])

plt.plot(t, square_wave)
```
